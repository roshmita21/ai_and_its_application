```python
numeric_df = df_copy.select_dtypes(include='number')
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

#Correlation Heatmap of Numeric Columns
Filters only numeric columns using select_dtypes(include='number').

Calculates pairwise correlation coefficients using .corr().

sns.heatmap() visualizes correlations with color gradients:

Red (close to 1): strong positive correlation

Blue (close to -1): strong negative correlation

Near 0: weak or no correlation

```python
sns.kdeplot(data=df_copy, x='Score')
plt.title('KDE of Scores')
plt.show()
```

#KDE Plot of Anime Scores

```python
q1 = df_copy['Score'].quantile(0.25)
q3 = df_copy['Score'].quantile(0.75)
iqr = q3 - q1
df_copy = df_copy[(df_copy['Score'] >= q1 - 1.5 * iqr) & (df_copy['Score'] <= q3 + 1.5 * iqr)]
```

#Removing Outliers from Score Using IQR Method
.quantile() method can calculate quartiles by specifying appropriate values (0.25, 0.5, 0.75).

```python
sns.boxplot(data=df_copy, x='Type', y='Score')
plt.title('Score by Anime Type')
plt.xticks(rotation=45)
plt.show()
```
