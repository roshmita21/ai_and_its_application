Doughnout chart

```python
import matplotlib.pyplot as plt

labels = ['Apples', 'Bananas', 'Mango', 'Litchi','Papaya']
sizes = [12,30, 25, 25, 20]
colors = ['pink','red', 'yellow', 'green', 'blue']

# Doughnut chart banayeko
plt.pie(sizes, labels=labels, colors=colors, startangle=270, wedgeprops={'width': 0.5})
plt.axis('equal')  # circular banauna
plt.title('Fruit Distribution Doughnut Chart')
plt.show()

```

```python
import matplotlib.pyplot as plt

labels = ['Apples', 'Bananas', 'Mango', 'Litchi','Papaya']
sizes = [12,30, 25, 25, 20]
colors = ['pink','red', 'yellow', 'green', 'blue']

# pie chart banayeko
plt.pie(sizes, labels=labels, colors=colors, startangle=270, wedgeprops={'width': 1.0})
plt.axis('equal')  # circular banauna
plt.title('Fruit Distribution Doughnut Chart')
plt.show()
```

```python
# Task 1: Create a Pandas Series with values [15, 25, 35, 45]. Calculate and print
# the mean.
import pandas as pd
data = pd.Series ([15,25,35,45])
mean = data.mean()
print("The mean is : ",mean)
```

```python
# Task 2: Create a Pandas Series with values [12, 8, 16, 4]. Calculate and print
# the median.
import pandas as pd
data = pd.Series ([12,8,16,4])
median = data.median()
print("The median is : ",median)
```

```python
# Task 3: Create a Pandas Series with values [2, 4, 2, 6, 4, 2]. Calculate and print
# the mode.
import pandas as pd
data = pd.Series ([2,4,2,6,4,2])
mode = data.mode()
print("The mode is : ",mode)
```

```python
# Task 4: Create a Pandas Series with values [7, 14, 21, 28]. Calculate and print
# the range.
import pandas as pd
data = pd.Series([7,14,21,28])
range = max(data)-min(data)
print("The range of data is:"range)
```

```python
# Task 5: Create a Pandas Series with values [5, 10, 15, 20]. Calculate and print
# the variance and standard deviation.
import pandas as pd
data = pd.Series([5,10,15,20])
variance = data.var()
standard_deviation = data.std()
print("The variance is :",variance)
print("The standard deviation is :",standard_deviation)
```
