### Exercise 4: Raading and processing a csv file

```python
#Using io module for input/output operations
import io

#Creating a csv file
csv_data = """Name,Grade
Arpan,89
Rupesh,68
Sabina,67
"""
#Reading a csv file

df = pd.read_csv(io.StringIO(csv_data))
print(df)


#Calculate average grade
average_grade = df['Grade'].mean()
print("Average grade:", average_grade)

#Calculate max grade
max_grade = df['Grade'].max()
print(f"Max Grade: {max_grade}")
```

## Exercise 5: Basic plotting with matplotlib

```python
import matplotlib.pyplot as plt
y = [5,4,3.25,3,3.25,4, 5]
x = [1,3,5,8,11,13, 15]

plt.plot(x, y, marker = 'o', color = 'red')

#using .title() function
plt.title("Line PLot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
# plt.
```

```python
plt.scatter(x, y)
```

## Exercise 5: Visualizing a dataframe with matplotlib

```python
import pandas as pd
df = pd.DataFrame({"Name":["Alice", "Bob", "Cathy", "Dina", "Elena", "Frank", "Gina"],
                   "Score":[34, 56, 78, 35, 89, 92,45]})

print(df)

plt.bar(df['Name'], df['Score'])
plt.title("Bar Graph")
plt.xlabel("Name")
plt.ylabel("Score")
plt.show()
# df['Score'].hist()
```
