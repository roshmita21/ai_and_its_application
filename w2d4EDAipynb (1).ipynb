```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#loading the dataset
df = sns.load_dataset("iris")
print(df.head())
print("Basic info")
print(df.info())
```

```python
#summary stats
print("Summary statistics")
print(df.describe())
```

```python
#checking null values
print("Checking null values")
print(df.isnull().sum())
```

```python
#duplicate rows
print("Checking duplicate row")
print(df.duplicated().sum())
```

```python
#value count on a categorical column
print(df["species"].value_counts())
print(df["petal_length"].value_counts())
```

```python
#Visualizing Categorical Variable
plt.figure(figsize= (6,8))
sns.countplot(x = "species",data =df)
plt.xlabel("species")
plt.ylabel("count of species")
```

```python
#Basic stats with functions
iris_numeric = df[["sepal_length","sepal_width","petal_length","petal_width"]]
print("Mean of each column")
print(iris_numeric.mean())
print("Median of each column")
print(iris_numeric.median())
print("Mode of each column")
print(iris_numeric.mode())
```

```python
#pair plot
sns.pairplot(df, hue = "species")
plt.show()
# In this figure different species is represented by different colors
#
```

```python
#correlation matrix (Numeric only) #correlation dui ota identity bich ko reln
corr = df.corr (numeric_only= True)
sns.heatmap(corr,annot = True , cmap = "coolwarm") #cmap =  Blues, Reds
plt.title("Correlation Matrix")
plt.show()

#This fig shows the correlation between different columns
#blue color indicated the negative corelation i.e increase of one means decrease of another
#red color indicates the positive corelation i.e when one increases another also increases and when one decreases another also decreases
```

```python
#creating a bloxplot
sns.boxplot(df)
plt.title("Boxplot")
plt.show()

#sepal_length has highest value of about 8cm
# petal_width has lowest values of about 0cm
#only sepal_width has outliers
#petal_length has the most variation
```

```python
#histogram
sns.histplot(df["petal_length"])
plt.show()

#The following figure shows count of petal length
#the count of petal is highest about 43 in between 1 abd 2 petal_length
#There is no count of petal between 2 and 3
```
