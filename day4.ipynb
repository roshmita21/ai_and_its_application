## File handline and exception

### Opening and Reading file

```python
##using open() and "r" to open and read a file
f = open("sample.txt", "r")
print(f.read())
f.close() ##good practice to close opened file
```

```python
##using open() and "r" to open and read a file
f = open("sample.txt", "r")
print(f.readline()) ##Reads the first line
print(f.readline()) ##Reads second line
f.close() ##good practice to close opened file
```

```python
f = open("sample.txt", "r")
print(f.read(8)) ##Opens first eight characters from the txt file
f.close()
```

```python
with open("sample.txt", "r") as file:
  print(file.read())

#Does not require a close function
```

### Writing to a file

```python
f = open("sample.txt", "w")
f.write("Hello Ruby!") #Overwrites the existing text to add this line
f.close()
```

```python
#For non existing file
f = open("new.txt", "w")
f.write("Creating a file that does not exist.")
f.close()
```

### Appending to a file

```python
f = open("new.txt", "a")
f.write("\nAppending a new line.") ##Adding a new line to existing file rather than overwriting
f.close()
```

### Basic exception handling


```python
print(a)

##Shows an error because a is not defined
```

```python
try:
  print(a)
except:
  print("Error!")

#Shows a custom error when a is not defined (for any kind of error-- not specific error)
```

```python
try:
  print(a)
except NameError:
  print("Variable 'a' is not defined.")
except TypeError:
  print("An error occured!")
```

```python
a = 333

try:
  print(a)
except NameError:
  print("Variable 'a' is not defined.")
except TypeError:
  print("An error occured!")
```

```python
try:
  x = int(input("Enter a number:")) #input has to be an int or we get an error
  print(10/x)
except ZeroDivisionError:
  print("Cannot divide by 0!")
except ValueError:
  print("Invalid input!")
finally:
  "Execution complete"
```

```python
try:
  with open("input.txt", "r") as input_file:
    input_lines = input_file.readlines()
    print("Lines in the input file:")
    print(input_lines)

  with open("output.txt", "w") as output_file:
    print("Lines in output file:")
    for line in input_lines:
      output_file.write(line)

  with open("output.txt", "r") as read_output:
    output_lines = read_output.read()
    print(output_lines)

except FileNotFoundError:
  print("Your input file does not exist.")
except:
  print("Error with code.")
```

```python
try:
  with open("data.txt", "r") as file:
    lines = file.readlines()
    num_lines = len(lines)

    if num_lines == 0:
      print("No lines in the data.txt file.")

  with open("line_count.txt", "w") as len_file:
    len_file.write(f"Total lines: {num_lines}")
    print(f"Total lines in data.txt: {num_lines}")
except FileNotFoundError:
  print("No file called data.txt found.")
```
